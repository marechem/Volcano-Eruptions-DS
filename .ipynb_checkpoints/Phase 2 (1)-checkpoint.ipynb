```python
# Read in all csv files into variables using pandas module
Events = pd.read_csv('events.csv')
volcano = pd.read_csv('volcano.csv')
```

```python
# Step 1: Encoding

nominal_features = ['primary_volcano_type', 'country', 'region', 'subregion', 'tectonic_settings', 'evidence_category', 'major_rock_1', 'major_rock_2', 'major_rock_3', 'major_rock_4', 'major_rock_5', 'minor_rock_1', 'minor_rock_2', 'minor_rock_3', 'minor_rock_4', 'minor_rock_5']

order_levels = {
    "tectonic_settings": ['Unknown', 'Subduction zone / Crustal thickness unknown', 'Subduction zone / Continental crust (>25 km)', 'Subduction zone / Intermediate crust (15-25 km)', 'Subduction zone / Oceanic crust (< 15 km)', 'Intraplate / Continental crust (>25 km)', 'Intraplate / Intermediate crust (15-25 km)',  'Intraplate / Oceanic crust (< 15 km)', 'Rift zone / Continental crust (>25 km)', 'Rift zone / Intermediate crust (15-25 km)', 'Rift zone / Oceanic crust (< 15 km)'],
    "evidence_category": ['Unrest / Holocene', 'Evidence Uncertain', 'Evidence Credible', 'Eruption Dated', 'Eruption Observed', ]
}

def encode(df):
    for col in nominal_features:
        df[col] = df[col].astype("category")
        
    for col, order_level in order_levels.items():
        df[col] = df[col].astype(CategoricalDtype(order_level, ordered=True))
        
    return df
        
    
volcano = encode(volcano) 
```

```python
# Step 2: Imputation

def drop_missing_values(df):
    df = df[df.columns[df.isnull().mean()<0.7]]
    df = df.loc[df.isnull().mean(axis=1)<0.7]
    
    return df

def impute(df):
    for col in df.select_dtypes("number"):
#         df[col] = df[col].fillna(0)
        df[col] = df[col].fillna(df[col].median())
    
    for col in df.select_dtypes("category"):
        if "None" not in df[col].cat.categories: df[col].cat.add_categories("None", inplace=True)
        df[col] = df[col].fillna("None")
        
    return df

volcano = drop_missing_values(volcano)
volcano = impute(volcano)

Events = drop_missing_values(Events)
Events = impute(Events)
```

```python
# Step 3: Handling Outliers

quant_names = ['population_within_5_km', 'population_within_10_km', 'population_within_30_km', 'population_within_100_km']

for col in quant_names:
    mean_val = volcano[col].mean()
    sd_val = volcano[col].std()
    lo = mean_val - 3*sd_val
    hi = mean_val + 3*sd_val
    volcano = volcano[(volcano[col] < hi) & (volcano[col] > lo)]
```

```python
event_count = Events.groupby(['volcano_number']).count()
event_count.rename(columns = {'volcano_name':'event_count'}, inplace = True)
event_count = event_count.drop(['eruption_number', 'eruption_start_year', 'event_number', 'event_type'], axis=1)
event_count
```

```python
volcano = pd.merge(volcano, event_count, on = "volcano_number", how = "inner")
volcano
```

```python
X1 = volcano[['latitude', 'event_count']].values
inertia_list1 = []

for n in range(1, 14):
    model = KMeans(n_clusters=n)
    model.fit(X1)
    inertia_list1.append(model.inertia_)
```

```python
plt.figure(1, figsize = (12, 6))
plt.plot(np.arange(1, 14), inertia_list1, 'o')
plt.plot(np.arange(1, 14), inertia_list1, '-', alpha = 0.5)
plt.xlabel('Total Clusters'), plt.ylabel('Inertia Value')
plt.show()
```

```python
model1 = KMeans(n_clusters=4)
model1.fit(X1)

labels1=model1.labels_

for c in set(labels1):  
    plt.scatter(X1[model1.labels_ == c, 0],
                X1[model1.labels_ == c, 1],
                label='Cluster {}'.format(c))
plt.xlabel('latitude')
plt.ylabel('event_count')
plt.legend()
plt.show()
```

```python
X2 = volcano[['longitude', 'event_count']].values
inertia_list2 = []

for n in range(1, 14):
    model = KMeans(n_clusters=n)
    model.fit(X2)
    inertia_list2.append(model.inertia_)
```

```python
plt.figure(1, figsize = (12, 6))
plt.plot(np.arange(1, 14), inertia_list2, 'o')
plt.plot(np.arange(1, 14), inertia_list2, '-', alpha = 0.5)
plt.xlabel('Total Clusters'), plt.ylabel('Inertia Value')
plt.show()
```

```python
model2 = KMeans(n_clusters=6)
model2.fit(X2)

labels2=model2.labels_


for c in set(labels2):  
    plt.scatter(X2[model2.labels_ == c, 0],
                X2[model2.labels_ == c, 1],
                label='Cluster {}'.format(c))
plt.xlabel('longitude')
plt.ylabel('event_count')
plt.legend()
plt.show()
```

```python
lat_cluster = pd.DataFrame({'lat_cluster': model1.labels_})
df_new = pd.concat([volcano, lat_cluster], axis=1, sort=False)
df_new = df_new[['country', 'latitude', 'longitude', 'lat_cluster', 'event_count']]
df_new.head(5)
```

```python
df_country = pd.get_dummies(df_new['country'])
df_new = df_new.drop(['country'], axis=1)
df_new = pd.concat([df_new, df_country], axis=1, sort=False)
df_new.head()
```

```python
decision_tree = DecisionTreeClassifier()
model_decision_tree = decision_tree.fit(X_train, y_train)
y_pred_decision_tree = model_decision_tree.predict(X_test)
y_pred_decision_tree
```

```python
print("Decision Tree Scores -->")
print("Accuracy: {}".format(round(accuracy_score(y_test, y_pred_decision_tree), 2)))

print("F1 Score: {}".format(round(f1_score(y_test, y_pred_decision_tree, average='weighted'), 2)))
print("Recall Score: {}".format(round(recall_score(y_test, y_pred_decision_tree, average='weighted'), 2)))
print("Precision Score: {}".format(round(precision_score(y_test, y_pred_decision_tree, average='weighted'), 2)))
```
